## 1. Estructura de Datos

Cada fragmento de lore tiene un identificador único, texto, categoría, película de origen y etiquetas.
Esta estructura permite búsqueda por categoría, por película y por ID.

```python
from dataclasses import dataclass, field
from typing import Optional

@dataclass
class LoreFragment:
    id: str
    text: str
    category: str
    film: str
    year_in_universe: Optional[str] = None
    tags: list = field(default_factory=list)

print("✓ Clase LoreFragment definida")
```

```python
def get_all_fragments():
    return LORE_FRAGMENTS

def get_fragments_by_category(category):
    return [f for f in LORE_FRAGMENTS if f.category == category.upper()]

def get_fragments_by_film(film):
    return [f for f in LORE_FRAGMENTS if film.upper() in f.film.upper()]

def get_fragment_by_id(fragment_id):
    return next((f for f in LORE_FRAGMENTS if f.id == fragment_id), None)

print("✓ Funciones de consulta definidas")
```

## 3. Explorando la Base de Datos

```python
from IPython.display import display, Markdown

categorias = ["PERSONAJE", "EVENTO", "OBJETO", "TIMELINE", "FACCION"]
tabla = "| Categoría | Fragmentos |\n|-----------|------------|\n"
for cat in categorias:
    n = len(get_fragments_by_category(cat))
    tabla += f"| {cat} | {n} |\n"

display(Markdown(f"**Total: {len(LORE_FRAGMENTS)} fragmentos**\n\n" + tabla))
```
